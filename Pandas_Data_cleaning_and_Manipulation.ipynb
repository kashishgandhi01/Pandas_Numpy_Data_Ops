3. IDENTIFYING OUTLIERS IN VARIOUS COLUMNS.


To determine the outliers, 1st quartile and 3rd quartile needs to be calculated, interquartile range = q3 - q1. With the help of which lower fence and upper fence of the data will be calculated, anything which lies outside of that range is considered as an outlier.

```python

Q1 = data_imputed_mean['Age'].quantile(0.25)
Q3 = data_imputed_mean['Age'].quantile(0.75)
IQR = Q3 - Q1

threshold = 1.5

outliers = data_imputed_mean[(data_imputed_mean['Age'] < Q1 - threshold * IQR) | (data_imputed_mean['Age'] > Q3 + threshold * IQR)]
outliers
```

Outliers in the Rating column are checked as follows and apparently none was found.

```python
Q1 = data_imputed_mean['Rating'].quantile(0.25)
Q3 = data_imputed_mean['Rating'].quantile(0.75)
IQR = Q3 - Q1

threshold = 1.5

outliers = data_imputed_mean[(data_imputed_mean['Rating'] < Q1 - threshold * IQR) | (data_imputed_mean['Rating'] > Q3 + threshold * IQR)]
outliers
```

4. LOCATION STANDARDISATION


```python
data_imputed_mean['Location'] = data_imputed_mean['Location'].replace({'New York,Ny':'New York', 'India,In':'India','India In':'India', 'Australia Aus':'Australia'})
data_imputed_mean
```

5. EXPLORING ESTABLISHMENTS COLUMN


```python
median = data_imputed_mean['Established'].median()
median
```

```python
data_imputed_mean['Established'] = data_imputed_mean['Established'].replace(-1, median)
print(data_imputed_mean)
```

6.
CHECKING THE EASY APPLY COLUMN

Some inconsistencies are found in the data which demand for the replacement of -1 with false in order to convert the column to boolean values only.


```python
data_imputed_mean['Easy Apply'] = data_imputed_mean['Easy Apply'].replace('-1','FALSE')
data_imputed_mean
```

7. ENSURING A CONSISTENT RATING SCALE

A consistent rating scale will ensure a more understandable rating, the scale is from 1 to 10.

```python
mapping = {-1:1, 0:1, 1:3, 2:4, 3:5, 4:6, 5:7, 6:8, 7:9, 8:10 }
data_imputed_mean['Rating'] = data_imputed_mean['Rating'].replace(mapping)
data_imputed_mean
```

8. FURTHER FORMATTING THE SALARY COLUMN

The salary column is of datatype obj which will be removed, and 3  columns for lower bound, upper bound and the average salary would be added to make our data more convinient to understand.

```python

data_imputed_mean[['Lower bound salary', 'Upper bound salary']] = data_imputed_mean['Salary'].str.split('-', expand=True)

data_imputed_mean['Lower bound salary'] = data_imputed_mean['Lower bound salary'].astype(int)
data_imputed_mean['Upper bound salary'] = data_imputed_mean['Upper bound salary'].astype(int)

data_imputed_mean['Average Salary'] = (data_imputed_mean['Lower bound salary'] + data_imputed_mean['Upper bound salary']) / 2

data_imputed_mean
```

```python

data_imputed_mean.drop(['Salary'], axis=1, inplace=True)
```
